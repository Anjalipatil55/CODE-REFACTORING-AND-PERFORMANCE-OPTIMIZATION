## ðŸ”´ Original Code (Before Refactoring)

```python

def process(data):
    total = 0
    for i in range(len(data)):
        if data[i] % 2 == 0:
            total = total + data[i]
    return total

def avg(data):
    s = 0
    for i in range(len(data)):
        s = s + data[i]
    return s / len(data)

def find_max(data):
    m = data[0]
    for i in range(len(data)):
        if data[i] > m:
            m = data[i]
    return m

numbers = [10, 15, 20, 25, 30, 35]

print("Even Sum:", process(numbers))
print("Average:", avg(numbers))
print("Maximum:", find_max(numbers))
```

## ðŸŸ¢ Refactored Code (After Optimization)
Improvements:
- Clear naming
- Built-in optimized functions
- Docstrings
- Type hints
- Better performance

```python

from typing import List

def sum_of_even_numbers(numbers: List[int]) -> int:
    """
    Calculates the sum of even numbers in the list.
    """
    return sum(num for num in numbers if num % 2 == 0)


def calculate_average(numbers: List[int]) -> float:
    """
    Calculates the average of numbers.
    Handles empty list safely.
    """
    if not numbers:
        return 0
    return sum(numbers) / len(numbers)


def find_maximum(numbers: List[int]) -> int:
    """
    Finds the maximum value in the list.
    """
    return max(numbers)


def main():
    numbers = [10, 15, 20, 25, 30, 35]
    print("Even Sum:", sum_of_even_numbers(numbers))
    print("Average:", calculate_average(numbers))
    print("Maximum:", find_maximum(numbers))

main()
```

### 3. Generator Expressions
- Reduced memory usage and improved performance

### 4. Error Handling
- Empty list handled safely in average calculation

### 5. Code Structure
- Introduced `main()` function
